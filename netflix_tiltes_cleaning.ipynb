```python
# Fill missing values in `director`, `cast`, and `country`
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df
```

```python
# Separate 'duration' column into 'duration_in_min' for movies and 'duration_in_seasons' for TV Shows.
df['duration_in_min'] = df[df['type'] == 'Movie']['duration'].str.extract('(\d+)').astype(float)
df['duration_in_seasons'] = df[df['type'] == 'TV Show']['duration'].str.extract('(\d+)').astype(float)

df
```

```python
# Fill missing values for `rating` and `date_added`
df['rating'].fillna('Unknown', inplace=True)
df['date_added'].fillna(method='ffill', inplace=True)

df
```

```python
# Convert `date_added` to datetime format with more flexible parsing
# Strip whitespace and handle mixed formats
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format='mixed', errors='coerce')

# If there are any NaT values after conversion, you can check them
# df[df['date_added'].isna()]

df
```

```python
# Handle misplaced duration values in the 'rating' column
incorrect_ratings = ['74 min', '84 min', '66 min']
df['rating'].replace(incorrect_ratings, 'Not Rated', inplace=True)

df
```

```python
# Print unique values again to confirm the changes
print("Unique values in 'rating' column after correction:")
print(df['rating'].unique())

df
```

```python
# Save the final cleaned DataFrame to a CSV file
df.to_csv(r"C:\Users\akars\OneDrive\Desktop\netflix_titles_fully_cleaned_final.csv", index=False)

df
```
